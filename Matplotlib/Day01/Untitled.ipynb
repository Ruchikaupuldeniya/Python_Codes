```python
### Add Label
```

```python

x=[1,2,3,4,5,6]
y=[4,8,9,4,5,8]
plt.title('Our First Graph With Variables',fontdict={'fontname':'Comic Sans MS','fontsize':20})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.plot(x,y)
plt.xticks([0,1,2,3])
plt.xticks([0,2,4,6,8,10])
plt.show()
```
